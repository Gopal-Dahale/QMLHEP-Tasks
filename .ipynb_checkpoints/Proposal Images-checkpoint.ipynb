```python
import h5py
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filename = "./data/jet_images/jet-images_Mass60-100_pT250-300_R1.25_Pix25.hdf5"

with h5py.File(filename, "r") as f:
    images = list(f['image'])
    labels = list(f['signal'])
```

```python
len(images), len(labels)
```

```python
images[0].shape
```

```python
boson = -1
qcd = -1
for index, i in enumerate(labels):
    if (i == 1) and (boson == -1):
        boson = index
    elif (i == 0) and (qcd == -1):
        qcd = index
    if (boson != -1) and (qcd != -1):
        break

print(boson, qcd)
```

```python
plt.imshow(images[boson], cmap = 'gray');
plt.title('W Boson')
plt.show()
```

```python
plt.imshow(images[qcd], cmap = 'gray');
plt.title('QCD')
plt.show()
```

```python
filename = "./data/EC_images/eplus.hdf5"

with h5py.File(filename, "r") as f:
    layer0 = list(f['layer_0'])
    layer1 = list(f['layer_1'])
    layer2 = list(f['layer_2'])
```

```python
fig = plt.figure(figsize=(40, 40))
columns = 3
rows = 1
for i in range(1, columns*rows +1):
    if i == 1:
        img = layer0[0]
    if i == 2:
        img = layer1[0]
    if i == 3:
        img = layer2[0]
    fig.add_subplot(rows, columns, i)
    plt.axis('off')
    plt.imshow(img,cmap='gray')
plt.show()
```
